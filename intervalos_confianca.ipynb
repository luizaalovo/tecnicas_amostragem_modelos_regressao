```python
import math
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

1. Se uma amostra aleatória n=25, tem uma média amostral de 51,3 e uma desvio padrão populacional de σ=2. Construa o intervalo com 95% de confiança para a média populacional µ.

```python
x_bar = 51.3  # média amostral
sigma = 2     # desvio padrão populacional
n = 25        # tamanho da amostra
z = 1.96      # valor crítico para 95% de confiança

erro_padrao = sigma / math.sqrt(n)
margem_erro = z * erro_padrao

ic_inferior = x_bar - margem_erro
ic_superior = x_bar + margem_erro

plt.figure(figsize=(6, 4))
plt.plot([ic_inferior, ic_superior], [1, 1], color='blue', lw=6, label="Intervalo de Confiança")
plt.scatter(x_bar, 1, color='red', zorder=5, label="Média Amostral")
plt.axvline(x=x_bar, color='red', linestyle='--', label="Média Amostral")
plt.title("Intervalo de Confiança para a Média Populacional")
plt.xlabel("Valor")
plt.yticks([])
plt.legend()
plt.show()

(ic_inferior, ic_superior)
```

2. Sabe-se que a vida em horas de um bulbo de lâmpada de 75W é distribuída de forma aproximadamente normal com desvio padrão de σ=25. Uma amostra aleatória de 20 bulbos tem uma vida media de 1.014 horas. Construa um intervalo de confiança de 95% para a vida média.

```python
x_bar_bulb = 1014  # média amostral
sigma_bulb = 25    # desvio padrão populacional
n_bulb = 20        # tamanho da amostra
z_bulb = 1.96      # valor crítico para 95% de confiança

erro_padrao_bulb = sigma_bulb / math.sqrt(n_bulb)
margem_erro_bulb = z_bulb * erro_padrao_bulb
ic_inferior_bulb = x_bar_bulb - margem_erro_bulb
ic_superior_bulb = x_bar_bulb + margem_erro_bulb

plt.figure(figsize=(6, 4))
plt.plot([ic_inferior_bulb, ic_superior_bulb], [1, 1], color='blue', lw=6, label="Intervalo de Confiança")
plt.scatter(x_bar_bulb, 1, color='red', zorder=5, label="Média Amostral")
plt.axvline(x=x_bar_bulb, color='red', linestyle='--', label="Média Amostral")
plt.title("Intervalo de Confiança para a Vida Média do Bulbo")
plt.xlabel("Horas")
plt.yticks([])
plt.legend()
plt.show()

(ic_inferior_bulb, ic_superior_bulb)
```

3. Qual deve ser o tamanho da amostra para que o intervalo com 99,5% de confiança para a média populacional tenha uma semi amplitude não superior a 1,5? Sabe-se que a variância populacional é de 23.

```python
Z = 2.576  # valor crítico para 99,5% de confiança
sigma = math.sqrt(23)  # desvio padrão populacional
E = 1.5  # semi-amplitude desejada

n = (Z * sigma / E) ** 2
math.ceil(n)  # arredondar para cima
```

4. Calcular o intervalo de confiança de 95% para a seguinte amostra, com variância populacional desconhecida:   19,8  18,5 17,6 16,7 15,8 15,4 14,1 13,6 11,9 11,4 11,4 8,8 7,5 15,4 15,4 19,5 14,9 12,7 11,9 11,4 10,1 7,9

```python
sample_data = [19.8, 18.5, 17.6, 16.7, 15.8, 15.4, 14.1, 13.6, 11.9, 11.4, 11.4, 8.8, 7.5, 15.4, 15.4, 19.5, 14.9, 12.7, 11.9, 11.4, 10.1, 7.9]

x_bar_sample = np.mean(sample_data)
s_sample = np.std(sample_data, ddof=1)
n_sample = len(sample_data)

t_value = stats.t.ppf(1 - 0.025, df=n_sample - 1)

erro_padrao_sample = s_sample / math.sqrt(n_sample)
margem_erro_sample = t_value * erro_padrao_sample
ic_inferior_sample = x_bar_sample - margem_erro_sample
ic_superior_sample = x_bar_sample + margem_erro_sample

plt.figure(figsize=(6, 4))
plt.plot([ic_inferior_sample, ic_superior_sample], [1, 1], color='blue', lw=6, label="Intervalo de Confiança")
plt.scatter(x_bar_sample, 1, color='red', zorder=5, label="Média Amostral")
plt.axvline(x=x_bar_sample, color='red', linestyle='--', label="Média Amostral")
plt.title("Intervalo de Confiança com Variância Desconhecida")
plt.xlabel("Valor")
plt.yticks([])
plt.legend()
plt.show()

(ic_inferior_sample, ic_superior_sample)
```

5. Uma marca particular de margarina diet foi analisada para determinar o nível em porcentagem de ácidos graxos insaturados. Uma amostra de seis pacotes resultou nos seguintes dados: 16,8; 17,2; 17,4; 16,9; 16,5 e 17,1. Encontre o intervalo de confiança de 99% para a amostra.

```python
margarine_data = [16.8, 17.2, 17.4, 16.9, 16.5, 17.1]

x_bar_margarine = np.mean(margarine_data)
s_margarine = np.std(margarine_data, ddof=1)
n_margarine = len(margarine_data)

t_margarine = stats.t.ppf(1 - 0.005, df=n_margarine - 1)

erro_padrao_margarine = s_margarine / math.sqrt(n_margarine)
margem_erro_margarine = t_margarine * erro_padrao_margarine
ic_inferior_margarine = x_bar_margarine - margem_erro_margarine
ic_superior_margarine = x_bar_margarine + margem_erro_margarine

plt.figure(figsize=(6, 4))
plt.plot([ic_inferior_margarine, ic_superior_margarine], [1, 1], color='blue', lw=6, label="Intervalo de Confiança")
plt.scatter(x_bar_margarine, 1, color='red', zorder=5, label="Média Amostral")
plt.axvline(x=x_bar_margarine, color='red', linestyle='--', label="Média Amostral")
plt.title("Intervalo de Confiança para Margarina Diet")
plt.xlabel("Porcentagem de Ácidos Graxos Insaturados")
plt.yticks([])
plt.legend()
plt.show()

(ic_inferior_margarine, ic_superior_margarine)
```

6. Uma amostra piloto com 12 elementos tece média de 6,7 e desvio padrão de 1,7. Qual deve ser o tamanho da amostra para que a semi amplitude do intervalo de 99,5% de confiança da média populacional não seja superior a 0,8 .

```python
Z = 2.576  # valor crítico para 99,5% de confiança
s = 1.7    # desvio padrão amostral
E = 0.8    # semi-amplitude desejada

n = (Z * s / E) ** 2
math.ceil(n)  # arredondar para cima
```

7. Em uma amostra aleatória de 85 mancais de eixos de manivelas de motores de automóveis, 10 têm um acabamento superficial mais rugoso do que as especificações permitidas. Calcule um intervalo de confiança para o 95% da proporção.

```python
p_hat = 10 / 85  # proporção de mancais com acabamento superficial mais rugoso
n_mancais = 85
z_mancais = 1.96  # valor crítico para 95% de confiança

erro_padrao_mancais = math.sqrt(p_hat * (1 - p_hat) / n_mancais)
margem_erro_mancais = z_mancais * erro_padrao_mancais
ic_inferior_mancais = p_hat - margem_erro_mancais
ic_superior_mancais = p_hat + margem_erro_mancais

plt.figure(figsize=(6, 4))
plt.plot([ic_inferior_mancais, ic_superior_mancais], [1, 1], color='blue', lw=6, label="Intervalo de Confiança")
plt.scatter(p_hat, 1, color='red', zorder=5, label="Proporção Amostral")
plt.axvline(x=p_hat, color='red', linestyle='--', label="Proporção Amostral")
plt.title("Intervalo de Confiança para Proporção de Mancais Rugosos")
plt.xlabel("Proporção")
plt.yticks([])
plt.legend()
plt.show()

(ic_inferior_mancais, ic_superior_mancais)
```

8. De 1.000 casos selecionados de aleatoriamente de câncer de pulmão, 823 resultaram em morte. Construa um intervalo de confiança de 95% para a taxa de morte de câncer de pulmão.

```python
p_death = 823 / 1000  # taxa de morte
n_cancer = 1000
z_cancer = 1.96  # valor crítico para 95% de confiança

erro_padrao_cancer = math.sqrt(p_death * (1 - p_death) / n_cancer)
margem_erro_cancer = z_cancer * erro_padrao_cancer
ic_inferior_cancer = p_death - margem_erro_cancer
ic_superior_cancer = p_death + margem_erro_cancer

plt.figure(figsize=(6, 4))
plt.plot([ic_inferior_cancer, ic_superior_cancer], [1, 1], color='blue', lw=6, label="Intervalo de Confiança")
plt.scatter(p_death, 1, color='red', zorder=5, label="Proporção Amostral")
plt.axvline(x=p_death, color='red', linestyle='--', label="Proporção Amostral")
plt.title("Intervalo de Confiança para Taxa de Morte de Câncer de Pulmão")
plt.xlabel("Taxa de Morte")
plt.yticks([])
plt.legend()
plt.show()

(ic_inferior_cancer, ic_superior_cancer)
```

9. Sabe-se que a proporção de animais contaminados com uma determinada doença não é superior a 10%. Qual deve o tamanho da amostra para que a semi amplitude do intervalo com 92% de confiança para a fração populacional não seja superior a 2%?

```python
Z = 1.75  # valor crítico para 92% de confiança
p = 0.1   # proporção máxima de animais contaminados
E = 0.02  # semi-amplitude desejada

n = (Z ** 2 * p * (1 - p)) / E ** 2
math.ceil(n)  # arredondar para cima
```
